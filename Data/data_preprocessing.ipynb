```python
#df_1m = pd.DataFrame()
df_500k = pd.DataFrame()

# List of years
years = [2018, 2019, 2020, 2021, 2022]

for year in years:
    df_year = df[df['year'] == year]
    
    #df_1m = pd.concat([df_1m, df_year.sample(n=200000, random_state=42)])
    
    df_500k = pd.concat([df_500k, df_year.sample(n=100000, random_state=42)])

#df_1m = df_1m.sample(frac=1, random_state=42).reset_index(drop=True)
df_500k = df_500k.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df_500k.head()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='zone_name', y='line_item_amount', data=df_500k)
plt.title('Sales Distribution by Zone')
plt.xlabel("Zone Name")
plt.ylabel("Line item amount")
plt.show()

# Discounts across zones
plt.figure(figsize=(10,6))
sns.violinplot(x='zone_name', y='bill_discount', data=df_500k)
plt.title('Discount Distribution by Zone')
plt.xlabel("Zone Name")
plt.ylabel("Bill Discount")
plt.show()

# Correlation matrix
corr = df_500k.corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Outliers detection
plt.figure(figsize=(10,6))
sns.boxplot(x='zone_name', y='line_item_amount', data=df_500k)
plt.title('Outliers in Sales by Zone')
plt.xlabel("Zone Name")
plt.ylabel("Line Item Amount")
plt.show()

# Sales trends over time by zone
df_500k['transaction_date'] = pd.to_datetime(df_500k['transaction_date'])
df_grouped = df_500k.groupby([df_500k['transaction_date'].dt.to_period('M'), 'zone_name'])['line_item_amount'].sum().unstack()
df_grouped.plot(figsize=(16, 8))
plt.xticks(ticks=pd.date_range(start='2018-01-01', end='2022-12-31', freq='MS'))
plt.title('Sales Trends Over Time by Zone')
plt.ylabel("Zone Name")
plt.xlabel("Transaction Date")
plt.legend()
plt.grid()
plt.show()
```
