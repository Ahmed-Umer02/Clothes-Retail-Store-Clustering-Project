# Data Preprocessing

```python
import pandas as pd
import pyarrow as pa
import pyarrow.csv as csv
from sklearn.preprocessing import LabelEncoder
```

```python
data_18 = 'data_csv/data_2018.csv'
data_19 = 'data_csv/data_2019.csv'
data_20 = 'data_csv/data_2020.csv'
data_21 = 'data_csv/data_2021.csv'
data_22 = 'data_csv/data_2022.csv'
```

```python
df_18 = pd.read_csv(data_18)
df_18.head()
```

```python
df_19 = pd.read_csv(data_19)
df_19.head()
```

```python
df_20 = pd.read_csv(data_20)
df_20.head()
```

```python
df_21 = pd.read_csv(data_21)
df_21.head()
```

```python
df_22 = pd.read_csv(data_22)
df_22.head()
```

```python
df = pd.concat([df_18, df_19, df_20, df_21, df_22], ignore_index=True)
df.head()
```

```python
df.describe()
```

```python
df.shape
```

### Handling Null values

The Dataset is too large to do One-Hot Encoding on my machine but that is the recommended method in this scenero to handle the null values.

Thus I am Impute missing values with 'Unknown'
I have also added in code for Impute missing values with the most frequent value However that is not in use in this project


```python
df.isnull().sum()
```

```python
# Impute missing values with 'Unknown'
df.fillna('Unknown', inplace=True)
df.head()
```

```python
# Impute missing values with the most frequent value
# for column in ['description','inventory_category','colour','size' ]:
#     most_frequent = df[column].mode()[0]
#     df[column].fillna(most_frequent, inplace=True)
# df.head()
```

```python
df.isnull().sum()
```

```python
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
df['line_item_amount'] = df['line_item_amount'].astype(float)
df['transaction_day'] = df['transaction_date'].dt.day
df['transaction_month'] = df['transaction_date'].dt.month
df.head()
```

```python
encoder = LabelEncoder()
df['description'] = encoder.fit_transform(df['description'])
df['inventory_category'] = encoder.fit_transform(df['inventory_category'])
df['colour'] = encoder.fit_transform(df['colour'])
df['size'] = encoder.fit_transform(df['size'])
df['zone_name'] = encoder.fit_transform(df['zone_name'])
df['store_name'] = encoder.fit_transform(df['store_name'])
df['size'] = encoder.fit_transform(df['size'])
df.head()
```

### Loading loading the data into a CSV within a zip file

```python
compression_options = dict(method='zip', archive_name='full_cleaned_data.csv')
df.to_csv('full_cleaned_data.zip', compression=compression_options)
```

### Load the data into a csv using pyarrow
when it comes to larger datasets it is much faster to load the data set with pyarrow instead of pandas


```python
new_df = pa.Table.from_pandas(df)
csv.write_csv(new_df, 'full_cleaned_data.csv')
```

```python
df_1m = pd.DataFrame()
df_500k = pd.DataFrame()

# List of years
years = [2018, 2019, 2020, 2021, 2022]

for year in years:
    df_year = df[df['year'] == year]
    
    df_1m = pd.concat([df_1m, df_year.sample(n=200000, random_state=42)])
    
    df_500k = pd.concat([df_500k, df_year.sample(n=100000, random_state=42)])

df_1m = df_1m.sample(frac=1, random_state=42).reset_index(drop=True)
df_500k = df_500k.sample(frac=1, random_state=42).reset_index(drop=True)
```
